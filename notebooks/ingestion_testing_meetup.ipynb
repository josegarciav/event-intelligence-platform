# Meetup Ingestion Testing

**Source**: Meetup GraphQL API  
**Type**: GraphQL API — requires `MEETUP_API_KEY`  
**Description**: Meetup is a platform for organising and finding local community events, professional meetups, hobby groups, and social gatherings. The Meetup GraphQL API requires an OAuth token or API key. This notebook tests the Meetup ingestion pipeline end-to-end.

> **Note**: Meetup requires a valid `MEETUP_API_KEY` (OAuth Bearer token) to access its GraphQL API at `https://api.meetup.com/gql`. Without the key the execute cell will fail at the fetch stage. Set the key in `services/api/.env` before running the full notebook.

## What this notebook does
1. Initialises the environment and verifies the `MEETUP_API_KEY` env var
2. Creates the Meetup pipeline via `PipelineFactory`
3. Executes the pipeline with a limited scope (max 10 events, 1 page) — requires API key
4. Inspects normalised `EventSchema` results
5. Explores `custom_fields` specific to Meetup
6. Saves the batch to `data/batches/meetup/`
7. Produces a DataFrame summary

```python
import sys
import os
import logging
from pathlib import Path

# Add services/api to path so src.* imports resolve
REPO_ROOT = Path("__file__").resolve().parents[1]
API_ROOT = REPO_ROOT / "services" / "api"
if str(API_ROOT) not in sys.path:
    sys.path.insert(0, str(API_ROOT))

# Load .env from services/api
try:
    from dotenv import load_dotenv
    load_dotenv(API_ROOT / ".env")
    print("dotenv loaded")
except ImportError:
    print("python-dotenv not installed — skipping .env load")

# Logging
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(name)s: %(message)s",
)
logger = logging.getLogger("notebook.meetup")

# Check required API key
# Meetup uses a GraphQL API (https://api.meetup.com/gql) with OAuth Bearer tokens.
# MEETUP_API_KEY should be set to a valid OAuth access token.
api_key = os.environ.get("MEETUP_API_KEY")
if api_key:
    print(f"MEETUP_API_KEY: set (length={len(api_key)})")
else:
    print("WARNING: MEETUP_API_KEY is not set.")
    print("Meetup uses OAuth — obtain a Bearer token from https://secure.meetup.com/oauth2/")
    print("Set it in services/api/.env or export it in your shell before running the execute cell.")
print(f"API_ROOT: {API_ROOT}")
```

```python
from src.ingestion.factory import PipelineFactory

factory = PipelineFactory()

# Temporarily enable the source for testing (disabled by default in YAML)
source_cfg = factory.get_source_config("meetup")
source_cfg["enabled"] = True

pipeline = factory.create_pipeline("meetup")
print(f"Pipeline created: {pipeline.__class__.__name__}")
print(f"Source type: {pipeline.source_type}")
print(f"Source name: {pipeline.config.source_name}")
```

```python
import asyncio

# NOTE: This cell requires MEETUP_API_KEY to be set.
# The Meetup GraphQL API (https://api.meetup.com/gql) uses OAuth Bearer tokens.
# Without a valid token the fetch stage will return an authentication error.
#
# What would happen with a valid key:
#   - Pipeline sends a GraphQL query to api.meetup.com/gql
#   - Fetches up to page_size=10 upcoming events for the configured location
#   - Normalises results into EventSchema objects with meetup-specific custom_fields
#     (e.g. group_name, group_urlname, rsvp_count, rsvp_limit, is_online_event)

result = await pipeline.execute(
    max_pages=1,
    page_size=10,
)

print(f"Status        : {result.status}")
print(f"Execution ID  : {result.execution_id}")
print(f"Total fetched : {result.total_events_processed}")
print(f"Successful    : {result.successful_events}")
print(f"Failed        : {result.failed_events}")
print(f"Duration      : {result.duration_seconds:.2f}s")
if result.errors:
    print(f"Errors        : {result.errors}")
```

```python
print(f"Inspecting {len(result.events)} normalised events\n")
print("-" * 70)

for i, event in enumerate(result.events):
    venue = event.location.venue_name or "N/A"
    city  = event.location.city or "N/A"
    start = event.start_datetime.isoformat() if event.start_datetime else "N/A"
    etype = event.event_type.value if event.event_type else "N/A"
    cf_keys = list(event.custom_fields.keys()) if event.custom_fields else []

    print(f"[{i+1:02d}] {event.title}")
    print(f"     Venue  : {venue} ({city})")
    print(f"     Date   : {start}")
    print(f"     Type   : {etype}")
    print(f"     CF keys: {cf_keys}")
    print()
```

```python
# Meetup-specific custom_fields typically include:
#   group_name       - the Meetup group hosting the event
#   group_urlname    - URL slug of the group (e.g. "python-london")
#   rsvp_count       - number of confirmed RSVPs
#   rsvp_limit       - maximum RSVPs allowed (if set)
#   waitlist_count   - number of people on the waitlist
#   is_online_event  - True if the event is virtual
#   event_hosts      - list of host member names
#   fee_amount       - event fee amount (if not free)
#   fee_currency     - fee currency code

print("custom_fields inspection for Meetup\n")
print("-" * 70)

for i, event in enumerate(result.events[:3]):
    print(f"Event: {event.title}")
    if event.custom_fields:
        for key, val in event.custom_fields.items():
            print(f"  {key}: {val}")
    else:
        print("  (no custom_fields)")
    print()

# Aggregate all custom_field keys across all events
all_cf_keys = set()
for event in result.events:
    if event.custom_fields:
        all_cf_keys.update(event.custom_fields.keys())
print(f"All custom_fields keys found: {sorted(all_cf_keys)}")
```

```python
from src.ingestion.batch_writer import BatchWriter

writer = BatchWriter()

batch_path = writer.save_batch(
    source_name="meetup",
    events=result.events,
    batch_id=result.execution_id,
    metadata={
        "pipeline_status": result.status.value,
        "duration_seconds": result.duration_seconds,
        "api_note": "Requires MEETUP_API_KEY OAuth Bearer token",
    },
)

print(f"Batch saved to : {batch_path}")
print(f"File size      : {batch_path.stat().st_size:,} bytes")
```

```python
import pandas as pd

if result.events:
    df = pipeline.to_dataframe(result.events)
    print(f"DataFrame shape: {df.shape}")
    print(f"\nColumns: {list(df.columns)}")
    print("\nBasic stats:")
    print(f"  Events          : {len(df)}")
    print(f"  Unique cities   : {df['city'].nunique()}")
    print(f"  Avg quality     : {df['data_quality_score'].mean():.2f}")
    print(f"  Free events     : {df['price_is_free'].sum()}")
    print("\nSample (title, city, start_datetime, data_quality_score):")
    print(df[["title", "city", "start_datetime", "data_quality_score"]].to_string(index=False))
else:
    print("No events to display — check that MEETUP_API_KEY is set and valid.")
```

```python
await pipeline.close()
print("Pipeline closed. Resources released.")
```
